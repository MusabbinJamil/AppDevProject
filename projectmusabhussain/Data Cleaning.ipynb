```python
df = final_project_data['date']

# Function to convert '731.0' to timedelta and then calculate the new date
def convert_to_date(value):
    days = int(float(value))  # Convert '731.0' to integer
    delta = timedelta(days=days)
    start_date = datetime(2015, 1, 8)  # Starting date
    new_date = start_date - delta
    return new_date

# Apply the function to the DataFrame column
df = df.apply(convert_to_date)
```
